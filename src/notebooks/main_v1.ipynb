## Reducción preliminar de features (TRAIN -> aplicar a TEST)
##### para eliminar constantes, missings altos e IDs / alta cardinalidad

```python
MAX_MISS_PCT = 40.0     
MAX_CARDIN_PCT = 60.0   
MAX_CAT_LEVELS = 200    

def preliminary_feature_reduction(X_tr, X_te):
    X_tr = X_tr.copy()
    X_te = X_te.copy()

    rep = data_report(X_tr)

    # 1) constantes (nunique<=1)
    const_cols = rep[rep["UNIQUE_VALUES"] <= 1].index.tolist()

    # 2) missings altos
    miss_cols = rep[rep["MISSINGS_%"] > MAX_MISS_PCT].index.tolist()

    # 3) IDs / cardinalidad altísima (casi único por fila)
    id_like_cols = rep[rep["CARDIN_%"] > MAX_CARDIN_PCT].index.tolist()

    # 4) categóricas con demasiados niveles
    cat_cols_local = X_tr.select_dtypes(exclude="number").columns
    high_levels = [c for c in cat_cols_local if X_tr[c].nunique(dropna=True) > MAX_CAT_LEVELS]

    drop_cols = sorted(set(const_cols + miss_cols + id_like_cols + high_levels))

    print("🔻 Drop constantes:", len(const_cols))
    print("🔻 Drop missings >", MAX_MISS_PCT, "%:", len(miss_cols))
    print("🔻 Drop ID-like (CARDIN% >", MAX_CARDIN_PCT, "%):", len(id_like_cols))
    print("🔻 Drop cat niveles >", MAX_CAT_LEVELS, ":", len(high_levels))
    print("🔻 TOTAL drop:", len(drop_cols))

    X_tr.drop(columns=drop_cols, inplace=True, errors="ignore")
    X_te.drop(columns=drop_cols, inplace=True, errors="ignore")

    return X_tr, X_te, drop_cols

X_train_red, X_test_red, dropped_cols = preliminary_feature_reduction(X_train, X_test)

print("✅ Shapes tras reducción preliminar")
print("Train:", X_train_red.shape, "| Test:", X_test_red.shape)

# Recalcular tipos tras reducción
num_cols = X_train_red.select_dtypes(include="number").columns.tolist()
cat_cols = [c for c in X_train_red.columns if c not in num_cols]
```

## Chequeo Outliers

```python
def quick_outlier_check(df_, cols, n=5):
    if len(cols) == 0:
        return
    print("🔎 Chequeo rápido de outliers (TRAIN) - top colas")
    for c in cols[:n]:
        s = df_[c]
        if s.dropna().empty:
            continue
        q1, q3 = np.nanpercentile(s, 25), np.nanpercentile(s, 75)
        iqr = q3 - q1
        if iqr == 0:
            continue
        lo, hi = q1 - 3*iqr, q3 + 3*iqr
        out_pct = ((s < lo) | (s > hi)).mean() * 100
        print(f"  - {c}: outliers(~3*IQR) = {out_pct:.2f}% (no se eliminan filas; se usa modelado robusto)")

quick_outlier_check(X_train_red, num_cols, n=8)
```

### Split train/test (estratificado)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=RANDOM_SEED,
    stratify=y
)

print("Train:", X_train.shape, "| Test:", X_test.shape)
```

### Preprocesado + Feature Engineering (pipeline)

```python
# numéricas: imputación median + escalado
# categóricas: imputación moda + one-hot
numeric_features = X_train.select_dtypes(include="number").columns.tolist()
categorical_features = [c for c in X_train.columns if c not in numeric_features]

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features),
    ],
    remainder="drop"
)

print("✅ Preprocessor creado")
```

### Baseline + comparativa de modelos (CV)
##### Métrica: F1 macro (buena con desbalanceo)

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)
scoring = {"f1_macro": "f1_macro", "accuracy": "accuracy"}

models = {
    "Dummy_most_frequent": DummyClassifier(strategy="most_frequent"),
    "LogReg": LogisticRegression(max_iter=3000),
    "RF": RandomForestClassifier(random_state=RANDOM_SEED),
    "GBC": GradientBoostingClassifier(random_state=RANDOM_SEED),
}

results = []
pipes = {}

for name, model in models.items():
    pipe = Pipeline(steps=[("preprocess", preprocessor), ("model", model)])
    pipes[name] = pipe

    cv_res = cross_validate(pipe, X_train, y_train, cv=cv, scoring=scoring, n_jobs=-1)
    results.append({
        "model": name,
        "f1_macro": float(np.mean(cv_res["test_f1_macro"])),
        "f1_macro_std": float(np.std(cv_res["test_f1_macro"])),
        "accuracy": float(np.mean(cv_res["test_accuracy"])),
        "accuracy_std": float(np.std(cv_res["test_accuracy"])),
    })

leaderboard = pd.DataFrame(results).sort_values("f1_macro", ascending=False)
print("📊 Leaderboard CV:")
display(leaderboard)
best_name = leaderboard.iloc[0]["model"]
print("🏆 Mejor baseline:", best_name)
```

### Hacemos Tuning del mejor modelo real

```python
if best_name == "Dummy_most_frequent":
    best_name = "LogReg"

best_pipe = pipes[best_name]
print("🔧 Modelo a tunear:", best_name)

if best_name == "LogReg":
    param_dist = {"model__C": np.logspace(-3, 2, 30)}
elif best_name == "RF":
    param_dist = {
        "model__n_estimators": [300, 600, 900],
        "model__max_depth": [None, 10, 20, 30],
        "model__min_samples_split": [2, 5, 10],
        "model__min_samples_leaf": [1, 2, 4],
    }
else:  # GBC
    param_dist = {
        "model__n_estimators": [150, 300, 500],
        "model__learning_rate": [0.01, 0.05, 0.1, 0.2],
        "model__max_depth": [2, 3, 4]
    }

search = RandomizedSearchCV(
    estimator=best_pipe,
    param_distributions=param_dist,
    n_iter=25,
    cv=cv,
    scoring="f1_macro",
    refit=True,
    random_state=RANDOM_SEED,
    n_jobs=-1,
    verbose=1
)

search.fit(X_train, y_train)
best_model = search.best_estimator_

print("Mejor F1_macro CV:", search.best_score_)
print("Mejores params:", search.best_params_)
```

### Evaluación final en TEST

```python
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

print("=== Classification report (TEST) ===")
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred, labels=sorted(y.unique()))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=sorted(y.unique()))
disp.plot(values_format="d", xticks_rotation=45)
plt.title("Confusion Matrix (TEST)")
plt.tight_layout()
savefig("confusion_matrix_test.png")
plt.show()
```
